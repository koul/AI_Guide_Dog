```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-8, 8)
xU, xL = x + 0.5, x - 0.5 
prob = ss.norm.cdf(xU, scale = 2.5) - ss.norm.cdf(xL, scale = 2.5)
prob = prob / prob.sum() # normalize the probabilities so their sum is 1
nums = np.random.choice(np.arange(0,16), size = 10000, p = prob)
plt.hist(nums, bins = len(x))
```
